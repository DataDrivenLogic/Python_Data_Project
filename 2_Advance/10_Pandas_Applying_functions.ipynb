Apply() - Calculate Projected Salary Next Year

```python
import pandas as pd
import matplotlib.pyplot as plt

#Loading data
df = pd.read_csv("hf://datasets/lukebarousse/data_jobs/data_jobs.csv")

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#Remove NaN values from column
df_salary = df[pd.notna(df['salary_year_avg'])].copy()

#Created a function to call the salary
def projected_salary(salary):
    return salary * 1.03

#using apply function, applied the function created above.
df_salary['salary_year_inflated']= df_salary['salary_year_avg'].apply(projected_salary)

df_salary[['salary_year_avg','salary_year_inflated']]
```

```python
## We can define in another way using anonymous function that is lambda function
df_salary['salary_year_inflated_2']= df_salary['salary_year_avg'].apply(lambda salary:  salary * 1.03)
df_salary[['salary_year_avg','salary_year_inflated_2']]
```

Lets take an another example of apply function. If we looked at the job_skills column which has list of skills but the column type is string. If we try to convert it directly using list() we will get error. So in this case we use ast.literal_eval() to convert it to list using the apply().
But, thing to remember here is that we need to remove the NaN values first.

```python
type(df['job_skills'][5])
```

```python
#We cant use directly the ast.literal_eval(), we need to define a function which would go through each row
import ast

def clean_list(skills_list):
    if pd.notna(skills_list):
        return ast.literal_eval(skills_list)

##Then use this function in apply function which will iterate through each row
df['job_skills'] = df['job_skills'].apply(clean_list)

```

```python
##Lets do using lambda function
df['job_skills'] = df['job_skills'].apply(lambda skills_list: ast.literal_eval(skills_list) if pd.notna(skills_list) else skills_list)
```

Now lets use apply() to "condition of" row. I.e. what if we want to apply a function on a row.
Lets assume inflated salary as below:

 Senior roles assume 5%
 Other roles assume 3%

```python
#Create a function to calculate this
#Created a function to call the salary
def projected_salary_roles(row):
    if 'Senior' in row['job_title_short']:
        return 1.05 * row['salary_year_avg']
    else:
        return 1.03 * row['salary_year_avg']

 
#using apply function applied the function created above.
#We have used here axis=1 which indicates we are apply these function on row.
#and we have directly use the apply function on dataframe "df_salary.apply(projected_s..." because
#With df_salary.apply(..., axis=1), each row (as a Series) is passed into your function → 
#so you can check row['job_title_short'] and row['salary_year_avg'].
df_salary['salary_year_inflated_roles']= df_salary.apply(projected_salary_roles, axis=1)
```
