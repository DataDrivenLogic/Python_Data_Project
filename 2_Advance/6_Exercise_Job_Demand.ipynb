```python
import pandas as pd
import matplotlib.pyplot as plt

#Loading data
df = pd.read_csv("hf://datasets/lukebarousse/data_jobs/data_jobs.csv")

#Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df['posted_month'] = df['job_posted_date'].dt.strftime('%B')
```

```python
#Created a new dataframe to check on the India job demands
df_India = df[df['job_country']=='India'].copy()
```

```python
df_India
```

```python
#Now using pivot table to get the posted_month column on index and on column side we are gonna use 
# job titles as we want tht counts of it. size in aggfunc is ised to get the count of rows
df_India_pivot = df_India.pivot_table(index='posted_month', columns='job_title_short', aggfunc='size')
df_India_pivot
```

```python
# Now if you see the index column i.e. posted_month is not in chronological order and we cant do the sorting by
#ascending or descending.
#For this we need to add a new column to sort months
#First we reset the ondex to push the months and then add month numbers so that we can sort by it

df_India_pivot.reset_index(inplace=True)

##NOTE: Remember each time you run reset_index() a new index is created and eventually will get error
        ##Best practice is to run each code in separate cell and if you write all code in same cell
        ##then run all from start
```

```python
#Created a new column to extract the month no. from the new posted_month column we created for sorting
df_India_pivot['month_no'] = pd.to_datetime(df_India_pivot['posted_month'], format='%B').dt.month
```

```python
#Sort it by month no
df_India_pivot.sort_values('month_no', inplace=True)
```

```python
##Once sorted,we are now setting the index with respect to the column we created posted_month
df_India_pivot.set_index('posted_month', inplace=True)
```

```python
##Drop the month_no column we created as we have sorted the posted_month and we dont need it
df_India_pivot.drop(columns='month_no', inplace=True)
```

```python
df_India_pivot
```

```python
##Now plotting the chart
df_India_pivot.plot(kind='line')
plt.legend()
```

```python
##Now, we are going to look at the top 3 job demands in india
##For this we create a new variable
top3 = df_India['job_title_short'].value_counts().head(3)

##We just want index so that we can use that in our chart converting it to list
top3 = top3.index.to_list()
```

```python
##Now we can use top3 in our pivot table to graph it
df_India_pivot[top3].plot(kind='line')
plt.legend()
plt.show()
```
