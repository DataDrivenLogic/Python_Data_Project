```python
monthly_counts = df['posted_month'].value_counts() ## Took the count of each month on the column 'posted month' we already had created

monthly_counts = monthly_counts.sort_index()

monthly_counts
```

##Bar Chart - Job Posting Counts

```python
job_counts = df['job_title_short'].value_counts()
job_counts
```

```python
job_counts = job_counts.sort_values(ascending=True) ##Sorting the values in ascending order, we used "job_counts = job_counts.sort_values()" to assign the sorted values back to job_counts

plt.barh(job_counts.index,job_counts) ##Horizontal bar chart as above we cant see the job titles
plt.show()
```

```python
##Matplotlib : Labelling
job_counts = job_counts.sort_values(ascending=False)
plt.bar(job_counts.index,job_counts)
plt.title('Postings By Job Title')
plt.ylabel('Count of Postings')
plt.xlabel('Month')
plt.xticks(rotation=45, ha='right') ## ha means horizontal alignment
plt.show()
```
